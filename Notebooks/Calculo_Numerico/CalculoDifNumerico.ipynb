
$$f^{'}(x_0)=\lim_{h\rightarrow 0} \frac{f(x_0+h)-f(x_0)}{h} $$

Una aproximación Natural es $$f^{'}(x_0)\approx \frac{f(x_0+h)-f(x_0)}{h}\quad (1)$$
para $h$ pequeño , pero cual es el error que se comete , que $h $ debemos seleccionar 

El teorema de Taylor nos ayuda  a evaluar ese error

$$f(x_0+h)=f(x_0)+f^{'}(x_0)h+\frac{h^2}{2}f^{''}(\xi)$$

donde $\xi$ es algún número entre $x_0$ y $x_0+h$

Ahora expresamos el error como:
$$f^{'}(x_0)-\frac{f(x_0+h)-f(x_0)}{h}=-\frac{h}{2}f^{''}(\xi) \quad (2)$$

$E_h=\mathcal O(h)$

Un estudio de refinamiento es el proceso de calcular una sequencia de aproximaciones mientras cambiamos $h$ con una cierta etrategia y observando la razón de dos errores consecutivos para segurarnos que el se va acercando al valor correcto en este caso nos mantenemos  tomado la mitad de  $h$ en cada paso para generar mas exactitud 
, pero mas precisamente los cocientes de los erroes sucesivos se aproximaran a 2 , en la practica un estudio de refinamiento es necesario para asegurarnos de que el método se ha implemnetado correctamente. es necesario ademas de que el error tienda a cero ,que las razones de errores sucesivos sean los correctos .  

¿Como seleccionar $h$?, uno pensaría que entre mas cercano este a 0 será mejor , también sabemos que hay un error de redondeo presente el cual al calcular la derivada es proporcional (aproximadamente ) a $\frac{1}{n}$, así el error total es de la forma $$E_n\approx cn+\frac{\delta}{n}$$  donde $\delta$  y $c$ son constantes La función de error tiene una forma como la de la grafica de abajo.Esto plantea severas restricciones en la exactitud que se puede alcanzar con esta formula  

image.png

Por ejemplo suponga que la unidad de máquina es $\epsilon\approx 10^{-16}$, será entonces que debemos tomar $h=10^{-16}$, de hecho no. Para ver esto notamos que nuestro error es $\mathcal O(h+\frac{\epsilon}{h})$, esta función es minimizada por $h\approx\sqrt{\epsilon}=10^{-8}$

Examinemos el impacto de los errores causados por el redondeo 
* $h$ pequeño $\Rightarrow \quad$  $f(x_0)$ y $f(x_0+h)$ tienen magnitudes similares y por lo tanto errores de redondeo similares supongamos que estas estan acotadas por $\delta$, y denote los valores calculados de $f\quad\text{por} \hat f$ el error total esta dado por

 $|f^{'}(x_0)-\frac{\hat f(x_0+h)-\hat f(x_0)}{h}|$  

  $\leq|f^{'}(x_0)-\frac{ f(x_0+h)- f(x_0)}{h}|+ |\frac{f(x_0+h)- f(x_0)}{h}- \frac{\hat f(x_0+h)-\hat f(x_0)}{h}|$  

  $\leq |\frac{h}{2}f^{''}(\xi)|+|\frac{ f(x_0+h)-\hat f(x_0+h)}{h}|+|\frac{ f(x_0)-\hat f(x_0)}{h}|$  

  $\leq |\frac{h}{2}f^{''}(\xi)|+\frac{2\delta}{h}$





Note que  $ |\frac{h}{2}f^{''}(\xi)|$ es el error de truncamiento en la aproximación de primera derivada . El segundo Termino de esta cota del error lo que provoca el crecimiento del error cuando $h\rightarrow 0$ un efecto similar se presenta en todas las formulas de aproximación para derivadas 

Enfoques similares se usan para aproximar derivadas de orden mas grande, por ejemplo la expansión de Taylor de $f(x_0\pm h)$ obtenemos

$$f^{''}(x_0)\approx \frac{f(x_0+h)-2f(x_0)+f(x_0-h)}{h^2}\quad (6)$$
cuyo error está dado por $ h^2f^{(4)}(\theta)/12$ . Similiarmente tenemos la fórmula  de 5 puntos
 $$f^{'''}(x_0)\approx \frac{-f(x_0+2h)+16f(x_0+h)-30f(x_0)+16f(x_0-h)-f(x_0-2h)}{12h^2}$$ que tien un eror  truncamiemto  de orden $h^4$.

 Los errores de redondeo se incrementan al csalcular derivadas de orden mayor, debido a que se incrementa la potencia reciproca de $h$ 

$$f(x_0+h)=f(x_0)+hf^{'}(x_0)+\frac{h^2}{2!}f^{''}(x_0)+\frac{h^3}{3!}f^{'''}(x_0)+\frac{h^4}{4!}f^{(4)}(x_0)+\cdots \quad \quad (7)$$

$$f(x_0-h)=f(x_0)-hf^{'}(x_0)+\frac{h^2}{2!}f^{''}(x_0)-\frac{h^3}{3!}f^{'''}(x_0)+\frac{h^4}{4!}f^{(4)}(x_0)+\cdots \quad \quad (8)$$



ahora sumando estas igualdades obtenemos :
$$f(x_0+h)+f(x_0-h)=2f(x_0)+h^2f^{''}(x_0)+\frac{h^4}{12}f^{(4)}(\theta)$$

despejando $f^{''}$ nos queda

### Ejemplo 


*   Considere un alambre de longitud $L$
*   Los extremos se mantiene con temperatura $0^{o}$C
*   $u(x,t)$ representa la temperatura en la posición $x$ y en el tiempo $t$
*   En $t=0$, se conoce $u(x,0)=u_0(x) $ para todo $x$
*   $$\rho c\frac{\partial u}{\partial t}-k\frac{\partial^2 u}{\partial x^2}=f(x,t)$$


>1.   $\rho$ es la densidad 
2.   $c$ es el calor específico 
3.   $f(x,t)$ es la fuente de calor 








queremos aproximar la distribución del calor a lo largo del alambre en tiempos especificos hasta cierto tiempo final $T_f$.

* $0 \leq t\leq T_f$ usando incrementos de $\Delta t$
* $0 \leq x\leq L$ usando incrementos de $\Delta x$

##### Notación
 Usamos $u_i^k\approx u(x_i,t^k)$, donde $t^k=k\Delta t \quad, x_i=i\Delta x$ con  $i=0,1,2,\cdots,n \quad\text{y}\quad  k=0,1,2,\cdots,m \quad$  para denotar la aproximación de la temperatura en esos  puntos 

### Aproximando el flujo de calor en un alambre aislado

$$u_t=\frac{u(x,t+\Delta t)-u(x,t)}{\Delta t}+\mathcal O(\Delta t)$$

Asi que:
$$u_t(x_i,t^k)\approx \frac{u_i^{k+1}-u_i^{k}}{\Delta t}$$

Para la segunda derivada usamos la ecuación (*)

$$u_{xx}=\frac{u(x-\Delta x,t)-2u(x,t)+u(x+\Delta x,t)}{\Delta x^2}+\mathcal O(\Delta x^2)$$
lo que con nuestra notación tenemos :
$$u_{xx}\approx \frac{u_{i-1}^k-2u_i^{k}+u_{i+1}^k}{\Delta x^2}$$

Juntando ambas aproximaciones podemos aproximar 
$\rho c\frac{\partial u}{\partial t}-ku_{xx}=f(x,t)$, con el modelo discreto 

$$\rho c [\frac{u_i^{k+1}-u_i^{k}}{\Delta t}]-K[\frac{u_{i-1}^k-2u_i^{k}+u_{i+1}^k}{\Delta x^2}]=f_i^k$$


dada la condición inicial $u(x,0)=u_0(x)$ y las condiciones de frontera se calcula $u_i^1\approx u(x_i,t^1) \quad \text{para}\quad i=0,1,2\cdots n$, ahora estos valores se usaran para calcular para aproximar las temperaturas en $t^2$ y asi sucesivamente.A esta formulación se le llama ** aproximación en difernecias finitas explicitas ** explicita significa que $u_i^{k+1}$ se calcula en terminos de $u_i^{k}$.

Sea $H=\frac{K\Delta t}{\rho\Delta x^2}$ y resolvamos para $u_i^{k+1}$ y así obtenemos 
$$u_i^{k+1}=Hu_{i-1}^k-(2H-1)u_{i}^k+Hu_{i+1}^k+\frac{\Delta t}{\rho c}f_i^k$$
