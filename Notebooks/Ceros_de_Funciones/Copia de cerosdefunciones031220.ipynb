**Método de la bisección**
. $[a,b]$ &  $f(a)*f(b)<0$

. tomamos $m=\frac{b+a}{2}$

. Si $f(m)*f(b)<0$ entonces $a=m$

. Si no $b=m$

. El proceso continua hasta que $(b-a)<\epsilon$

```python
 # defina la  funcion, la tolerancia y el intervalo con el que se empieza [a, b]
import numpy as np
import matplotlib.pyplot as plt
f = lambda x: np.exp(x) - 2
tol = 0.1
a, b = -2, 2
x = np.linspace(-2.1, 2.1, 1000)
# graph the function f
fig, ax = plt.subplots(1, 1, figsize=(12, 4))
ax.plot(x, f(x), lw=1.5)
ax.axhline(0, ls=':', color='k')
ax.set_xticks([-2, -1, 0, 1, 2])
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$f(x)$', fontsize=18)
# encuentre la raiz usando el método de biseccion y visualize
# los pasos en la visualización 
fa, fb = f(a), f(b)
ax.plot(a, fa, 'ko')
ax.plot(b, fb, 'ko')
ax.text(a, fa + 0.5, r"$a$", ha='center', fontsize=18)
ax.text(b, fb + 0.5, r"$b$", ha='center', fontsize=18)
n = 1
while b - a > tol:
    m = a + (b - a)/2
    fm = f(m)
    ax.plot(m, fm, 'ko')
    ax.text(m, fm - 0.5, r"$m_%d$" % n, ha='center')
    n += 1
    if np.sign(fa) == np.sign(fm):
            a, fa = m,fm
    else:
            b, fb = m, fm
ax.plot(m, fm, 'r*', markersize=10)
ax.annotate("Aproximación de la raiz  %.3f" % m,
fontsize=14, family="serif",
xy=(a, fm), xycoords='data',
xytext=(-150, +50), textcoords='offset points',
arrowprops=dict(arrowstyle="->",
connectionstyle="arc3, rad=-.5"))
ax.set_title("Bisection method")
```

**Método de Newton**

La fórmula de Newton se puede drivar de la formula de taylor $f(x)$ alrededor de $x$ 

.  $$f(x_{i+1})=f(x_i)+f^{'}(x_i)(x_{i+1}-x_i)+O(x_{i+1}-x_i)^2$$

.  $$0=f(x_i)+f^{'}(x_i)(x_{i+1}-x_i)+O(x_{i+1}-x_i)^2$$

. $$x_{i+1}=x_i-\frac{f(x_i)}{f^{'}(x_i)}$$

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
tol = 0.01
xk = 2
s_x = sy.symbols("x")
s_f = sy.exp(s_x) - 2
f = lambda x: sy.lambdify(s_x, s_f, 'numpy')(x)
fp = lambda x: sy.lambdify(s_x, sy.diff(s_f, s_x), 'numpy')(x)
x = np.linspace(-1, 2.1, 1000)
# setup a graph for visualizing the root finding steps
fig, ax = plt.subplots(1, 1, figsize=(12, 4))
ax.plot(x, f(x))
ax.axhline(0, ls=':', color='k')
# iterate Newton's method until convergence to the desired tolerance has been reached
n = 0
while f(xk) > tol:
    xk_new = xk - f(xk) / fp(xk)
    ax.plot([xk, xk], [0, f(xk)], color='k', ls=':')
    ax.plot(xk, f(xk), 'ko')
    ax.text(xk, -.5, r'$x_%d$' % n, ha='center')
    ax.plot([xk, xk_new], [f(xk), 0], 'k-')
    xk = xk_new
    n += 1
    ax.plot(xk, f(xk), 'r*', markersize=15)
ax.annotate("Root approximately at %.3f" % xk,
fontsize=14, family="serif",
xy=(xk, f(xk)), xycoords='data',
xytext=(-150, +50), textcoords='offset points',
arrowprops=dict(arrowstyle="->",
connectionstyle="arc3, rad=-.5"))
ax.set_title("Newtown's method")
ax.set_xticks([-1, 0, 1, 2])
plt.show()
```

Función para calcular la raíz de una función dando la aproximación inicial y su derivada 

```python
def newton(fcn,df,g,tol):
  old=0 #iniciar la interacion
  while abs(g-old)>tol:
    old=g
    g=old-fcn(old)/df(old)
  sol=g
```

Programa que combina el método de bisección con el de Newton

```python
# %load prg_NewtonR.py
"""
Created on Mon Jun 29 12:32:20 2020

@author: 
"""


## module newtonRaphson
''' root = newtonRaphson(f,df,a,b,tol=1.0e-9).
    encuentre una raíz  f(x) = 0  combinando el método de  Newton-Raphson
    con el de bisección. la Ráiz debe de estar contenida en  (a,b).
    el usuario debe de facilitar la funcion f(x) y su derivada  df(x).   
'''    
def newtonRaphson(f,df,a,b,tol=1.0e-9):
    import error
    from numpy import sign
    
    fa = f(a)
    if fa == 0.0: return a
    fb = f(b)
    if fb == 0.0: return b
    if sign(fa) == sign(fb): error.err('la raíz no está contenida ')
    x = 0.5*(a + b)                    
    for i in range(30):
        fx = f(x)
        if fx == 0.0: return x
      # Tighten the brackets on the root 
        if sign(fa) != sign(fx): b = x  
        else: a = x
      # Try a Newton-Raphson step    
        dfx = df(x)
      # If division by zero, push x out of bounds
        try: dx = -fx/dfx
        except ZeroDivisionError: dx = b - a
        x = x + dx
      # If the result is outside the brackets, use bisection  
        if (b - x)*(x - a) < 0.0:  
            dx = 0.5*(b - a)                      
            x = a + dx
      # Check for convergence     
        if abs(dx) < tol*max(abs(b),1.0): return x
    print('Demasiadas iteraciones  Newton-Raphson')
```

## El método de la secante 

. El método de la secante usa el punto donde la recta secante que pasa por las dos iteraciones anteriores $x_0,x_1$  corta el eje $x$

. La pendiente de la secante es $$\frac{f(x_1)-f(x_0)}{x_1-x_0}$$

.  la ecuación es $$y-f(x)=\frac{f(x_1)-f(x_0)}{x_0-x_1}(x-x_1)$$

. Tomando $y=0$ y resolviendo para $x$ obtenemos la siguiente iteración

$$x_2=x_1-\frac{x_1-x_0}{f(x_1)-f(x_0)}f(x_1)$$

. La expresión general es 

$$x_{n+1}=x_n-\frac{x_{n}-x_{n-1}}{f(x_n)-f(x_{n-1})}f(x_n)$$
