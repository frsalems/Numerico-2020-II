Problema general de encontrar un polinomio $p(x)=a_0+a_1x+a_2x^2+\cdots+a_Mx^M$ que ajuste lo datos $D=\{(x_1,y_1),(x_2,y_2),\dots,(x_N,y_N)\}$

El problema planteado es encontrar $a_0,a_1,\cdots,a_M$ tal que: $p(x_i)$ coincida lo mejor posible con los datos  

&nbsp;
$$
\min_a\sum_{i=1}^N (y_i-p(x))^2
$$
&nbsp;

El polinomio buscado debe de cumplir 
&nbsp;

$$
\begin{align}
p(x_1)&=a_0+a_1x_1+a_2x_1^2+a_3x_1^3+\cdots +a_Mx_1^M \\
p(x_2)&=a_0+a_1x_2+a_2x_2^2+a_3x_2^3+\cdots +a_Mx_2^M \\
p(x_3)&=a_0+a_1x_3+a_2x_3^2+a_3x_3^3+\cdots +a_Mx_3^M \\
       \vdots& \quad  \quad  \quad \vdots \quad  \quad  \quad \vdots \quad  \quad  \quad \vdots\\
p(x_N)&=a_0+a_1x_N+a_2x_N^2+a_3x_N^3+\cdots +a_Mx_N^M 
\end{align}
$$

### Matricialmente se puede escribir 
&nbsp;

$$
\begin{bmatrix}
1&x_1&x_1^2&x_1^3&\cdots& x_1^M \\
1&x_2&x_2^2&x_2^3&\cdots& x_2^M \\
1&x_3&x_1^2&x_3^3&\cdots& x_3^M \\
\vdots&\vdots&\vdots&\vdots& \cdots& \vdots\\
1&x_N&x_N^2&x_N^3&\cdots& x_N^M 
\end{bmatrix}
\begin{bmatrix}
a_0 \\
a_1 \\
a_2 \\
\vdots\\
a_M)
\end{bmatrix}
=
\begin{bmatrix}
p(x_1) \\
p(x_2) \\
p(x_3) \\
\vdots\\
p(x_M)
\end{bmatrix}
$$
&nbsp;


Recordemos que $Ax$ se puede escribir  como 
&nbsp;
$$Ax=b=x_1A_1+x_2A_2+\cdots+x_mA_m=b$$,
&nbsp;

Entonces para que el sistema $Ax=b$ tenga solución, $b$ tiene que ser una combinacion linea de las columnas de $A$

&nbsp;

$$
\min_a \sum_{i=1}^N(y_i-p(x_i))^2=(\mathbf{y}-Aa)^{t}(\mathbf{y}-Aa) 
$$
&nbsp;

$$x^tx=x\cdot x=\|x\|^2$$


### Nuestro problema se puede escribir como 

&nbsp;

$$
\min_a \sum_{i=1}^N(y_i-p(x_i))^2=(\mathbf{y}-Aa)^{'}(\mathbf{y}-Aa) 
$$
&nbsp;

Dicho de otra manera  necesitamos minimizar $$\|y-Aa\|$$ con respecto a $a$

$$A^TA\hat{x}=A^Tb$$

#### Lema 
El nucleo de $A$ tiene la misma dimension que el nucleo de $A^TA$

**DEM**

. si $x$ está en el nucleo de $A$ tenemos  $Ax=0$, entonces $A^TAx=0$ entonces $kernel(A)\subset kernel(A^TA)$ 

. si $x\in kernel(A^TA)$ entonces $A^TAx=0$ de aqui $x^tA^TAx=0$ por lo tanto $(Ax)^TAx=0$


**Nota ** si las columnas de $A$ son linealmente independientes entonces $A^TA$ es invertible

## Mínimos cuadrados continuos 

Encontrar el polinoimio de orden $m$ que mejor ajuste a los datos $(x_i,y_i)$ donde $i=1\cdots n $ 

 ie)que Minimize el error cuadrático 
$$\sum |(y_i-p(x_i)|^2$$
donde $p(x)=a_0+a_1x+a_2x^2+\cdots+a_mx^m$

El problema minimos cuadrados continuo, consiste en :

 1. Dada una colección de funciones $\{\phi_1,\phi_2,\cdots,\phi_n,\cdots\}$ definidas en $[a,b]$ cuadrado integrables ie) $$\int_a^b|\phi_i|^2dx<\infty$$ y una función $f$ cuadrado integrable 
 
 2. Encontrar una función $\omega$ en el subespacio generado por $\{\phi_,\phi_2,\cdots,\phi_n,\cdots\}$ es decir $\omega=a_0\phi_0+a_1\phi_1+\cdots+a_M\phi_m$
 
 3. Que minimize $$ L_2(f,\omega)=\|f-\omega\|=\sqrt{\int_a^b |f-\omega|^2dx}$$
 
 4. Nosotros trabajaremos con las funciones $\{1,x,x^2,\cdots,x^n,\cdots\}$

Así, nosotros pondermos $\omega=p$, y para simplificar la notacion pondremos

$$ L_2(f,p)^2=F(a_0,a_1\cdots,a_m)=F(\bf{a})$$

Claramente queremos que la selecion de $a_0,a_1,\cdots,a_m$  minimize $F$ también minimizará $L_2(f,p)$

Sean $\phi_0,\phi_1,\cdots,\phi_M$ polinomios de grados $0,1,\cdots,M$ repsectivamente , entonces estos polinomios forman una base para el espacio de polinomios de grado $\leq M$. El polinomio de aproximación $p$ puede escribirse como una combinación lineal de esta base de polinomios :
&nbsp;

$$
p(x)=a_0\phi_0(x)+a_1\phi_1(x)+\cdots+a_M\phi_M(x)
$$
&nbsp;

El problema se puede plantear como sigue : Minimizar
&nbsp;

$$
F(a_0,a_1,\cdots,a_M)=\int_a^b[f(x)-(a_0\phi_0(x)+a_1\phi_1(x)+\cdots+a_M\phi_M(x))]^2dx
$$
&nbsp;

Se aplica el mismo procedimiento que antes , seleccionamos que las parciales de $F$ sean 0 para obtener el sitema de ecuaciones para los coeficientes desconocidos
&nbsp;

$$
\begin{bmatrix}
c_{00}&c_{01}&c_{02}&\cdots&c_{0M} \\
c_{10}&c_{11}&c_{11}&\cdots& c_{1M}\\
c_{20}&c_{21}&c_{22}&\cdots&c_{2M} \\
\vdots&\vdots&\vdots&\vdots& \cdots& \vdots\\
c_{M0}&c_{M1}&c_{M2}&c_{M3}&\cdots& c_{MM}
\end{bmatrix}
\begin{bmatrix}
a_0 \\
a_1 \\
a_2 \\
\vdots\\
a_M
\end{bmatrix}
=
\begin{bmatrix}
b_0 \\
b_1 \\
b_2 \\
\vdots\\
b_M
\end{bmatrix} \quad (*)
$$
&nbsp;

Donde 
&nbsp;


$$
\begin{align}
c_{ij}&=c_{ji}=\int_a^b \phi_i(x)\phi_j(x) dx\\
b_i&=\int_a^b f(x)\phi_i(x)dx
\end{align} 
$$





La clave es encontrar  un conjunto de polinomios $\phi_0,\phi_1,\cdots,$ de tal suerte que cada $\phi_i$ sea de grado $i$ tal que 
&nbsp;

$$
c_{ij}=\int_a^b\phi_{i}(x) \phi_j(x)dx=0 \quad i\neq j
$$
&nbsp;

siempre que $i\neq j$  en este caso sistema (*) se reduce a un sistema diagonal 
&nbsp;

$$
\begin{bmatrix}
c_{00}&0&0&\cdots&0 \\
0&c_{11}&0&\cdots& 0\\
0&0&c_{22}&\cdots&0 \\
\vdots&\vdots&\vdots&\vdots&  \vdots\\
0&0&0&0& c_{MM}
\end{bmatrix}
\begin{bmatrix}
a_0 \\
a_1 \\
a_2 \\
\vdots\\
a_M
\end{bmatrix}
=
\begin{bmatrix}
b_0 \\
b_1 \\
b_2 \\
\vdots\\
b_M
\end{bmatrix}
$$
&nbsp;


cuya solución se puede obtener facilmente 
&nbsp;

$$
a_i=\frac{b_{j}}{c_{jj}}=\frac{\int_a^b f(x)\phi_j(x)dx}{\int_a^b [\phi_j(x)]^2} dx
$$
&nbsp;


A los polinomios con esta propiedad se les llama *polinomios ortogonales* sobre $[a,b]$.
&nbsp;

Nos concentraremos en un conjunto particular de polinomios llamados  los polinomios de Legendre  $P_n(x)$. si multiplicamos funciones ortogonales por escalares, no se afectará la ortogonalidad y algo de normalización será útil.

Para  los polinomios de Legendre seleccionaremos $P_n(1)=1$ para toda $n$ los primeros polinomios en el intervalo $[-1,1]$ son
&nbsp;

$$
\begin{align}
P_0(x)&=1,P_1(x)=x\\
P_2(x)&=(3x^2-1)/2\\
P_3(x)&=(5x^3-3x)/2\\
P_4(x)&=(35x^4-30x^2+3)/8\\
\end{align} 
$$


*Encuentre* los primeros tres polinomios ortogonales en el intervalo $[0,\pi]$ y uselos para encontrar la aproximación de minimos cuadrados del $seno x$ en este intervalo.

El primero ,$\phi_0$ tien grado cero y coeficiente (constante) $1$. A medida que vayamos encontrando los polinomios,los normalizaremos para que el coeficiente principal sea $1$. entonces 
&nbsp;

$$
\phi_0(x)=1
$$
&nbsp;

Enseguida $\phi_1(x)$ debe de ser de la forma $\phi_1(x)=x+c$ y debe de ser ortogonal $
\phi_0(x)
$

&nbsp;

$$
\int_0^{\pi}\phi_{1}(x) \phi_0(x)dx=\int_0^{\pi}(x+c) (1)dx=0
$$
&nbsp;

Se sigue que $\pi^2/2+c\pi=0$ entonces $c=-\pi/2$ y
&nbsp;
$$
\phi_1(x)=x-\frac{\pi}{2}
$$
&nbsp;

Ahora sea $\phi_2=x^2+ax+b$ debemmos escoger $a,b$ tal que sea ortogonal a $\:\phi_0,\:\text{y}\:\phi_1$ o loque eslo mismo forzar a $\phi_1(x)$ a ser ortogonal tanto a $1$ como a $x$ estas condiciones son
&nbsp;

$$
\begin{align}
\int_0^{\pi}(x^2+ax+b)dx&=\frac{\pi^3}{3}+\frac{a\pi^2}{2}+b\pi=0\\
\int_0^{\pi}x(x^2+ax+b)dx&=\frac{\pi^4}{4}+\frac{a\pi^3}{3}+\frac{b\pi^2}{2}=0
\end{align}
$$
&nbsp;


 De donde obtenemos $a=-\pi\: \text{y}\:b=\pi^2/6$,asi 
 &nbsp;
 
 $$
 \phi_2(x)=x^2-\pi x+\pi^2/6
 $$
 &nbsp;
 
 Los coeficientes del sistema son 
 &nbsp;
 
 $$
 \begin{align}
 c_{00}&=\pi,\quad c_{11}=\frac{\pi^3}{12},\quad c_{22}=\frac{\pi^5}{180}\\
 b_0&=2,\quad b_1=0,\quad b_2=\frac{\pi^2}{3}-4
 \end{align}
 $$
 &nbsp;
 
De donde obtenemos 
&nbsp;
 
 $$
 a_0=\frac{2}{\pi},\quad a_1=\frac{0}{\pi^3/12},\quad a_2=\frac{(\pi^2-12)/3}{\pi^5/180}
