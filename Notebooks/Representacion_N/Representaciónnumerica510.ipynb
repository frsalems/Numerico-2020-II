
## Representación numérica
Cualquier número positivo $x$ se puede expresar en teminos de una base $\beta$ como \\

$$x=a_n\beta^n+a_{n-1}\beta^{n-1}+\cdots+a_1\beta^1+a_0\beta^0+a_{-1}\beta^{-1}+a_{-2}\beta^{-2}+\cdots+a_{-m}\beta^{-m}$$

donde las $a_{i's}=0,1,\cdots,\beta-1$

234.23=$2*10^2+3*10^1+3*10^0+2*10^{-1}+3*10^{-2}$

$$x=\sum_{k=-m}^{n} a_k\beta^k$$

La forma de punto flotante de $x$

$$x=f\times\beta^E$$

donde la mantisa $f$ satisface

$$1\leq f \leq \beta $$

para aproximar la mantisa con solo un número fijo de cifras signficativas, por lo que $f$ toma la forma 

$$x=b_0+b_{1}\beta^{-1}+b_{2}\beta^{-2}+\cdots+a_{N}\beta^{-N}=\sum_{k=0}^{N} b_k\beta^{-k}$$

En el sistema binario $\beta=2$ por lo que $1\leq f <2$ , entonces $b_0=1$

Recuerde que las cifras significativas son todas aquellas que estan después de la primera cifra diferente de cero 

. los 0´s que se encuentran entre dos cifrasdiferentes de cero son significantes

. los ceros que van al principio no son significantes 

. los ceros que van al final se consideran significantes 


por ejemplo 1.2345 y 0.012345 ambos tienen 5 cifras significativas, mientras que 10.2345 y 0.0123450 tienen 6

### Aritmética de punto flotante 
. Un número positivo se puede representar de forma normalizada de la siguiente forma :
$$\hat x=\hat f\beta^E$$

donde $\hat f$ consiste de un número fijo de numeros (N+1),  de base $\beta$
$$\hat f=\hat d_0.\hat d_1\hat d_2\cdots\hat d_N \quad (1*)$$

donde $1\leq \hat d_0< \beta \quad$   y  $\quad 0\leq \hat d_k< \beta \quad$ para $\quad k=1,2,\dots,N$

El valor absoluto del error al representar $x=f\beta^E=\beta^E( d_0. d_1 d_2\cdots \quad )$ por $\hat x$ es:


$$|x-\hat x|=\beta^E|f-\hat f|\quad (2*)$$

Si $\hat f$ se obtiene po truncamiento entonces $\hat d_k=d_k \quad$ para $k=0,1,2,\cdots,N$ asi que (2*) resulta en

$$|x-\hat x|=\beta^E\sum_{k={N+1}}^\infty d_k\beta^{-k}\leq \beta^{E-N}$$

$$\sum_{k={N+1}}^\infty d_k\beta^{-k}=\beta^{-N}\sum_{1}^\infty  d_k\beta^{-k}\leq \beta^{-N}\sum_{1}^\infty (\beta-1)  \beta^{-k}=\beta^{-N}(\sum_{0}^\infty   \beta^{-k}-\sum_{1}^\infty   \beta^{-k})= \beta^{-N}$$


La primera desigualdad se cumple porque $d_k\leq\beta-1$

El Redondeo es equivalente a primero sumar $\beta /2 \quad \text{a}\quad d_{N+1}$ y después truncar obtener:

Para redondeo tenemos :
$$|x-\hat x|=\frac{\beta^{E-N}}{2} \quad (3*)$$

Para el error relativo tenemos:


$$\frac{|f-\hat f|}{|f|}\leq\left\{ \begin{array}{lcc}
             \frac{1}{\beta^N} &   truncamiento \\
             \\ \frac{1}{2\beta^N} & redondeo \\
             \end{array}
   \right. \quad (4*)$$ 

Para probar la cota para el truncamiento nótese que:
$$\frac{|f-\hat f|}{|f|}=\frac{\sum_{k={N+1}}^\infty d_k\beta^{-k}}{|f|}\leq \frac{\beta^{-N}}{|f|}\leq \beta ^{-N} $$

Una  normas de la IEEE es que los resultados almacenados de cualquier operación de punto flotante sea el redondeado del valor exacto, (es decir  se consideran como valores exactos)

  Esta Norma implica que el redondeo y normalización final no tenga un error relativo mayor a las cotas en (4*). Sin embargo esto no es la úinica fuente de error ya que los datos usualmente no son exactos. Si $\diamondsuit$ en $x\diamondsuit y$ representa una de las operaciones artimeticas $(+,- ,\times, \div )$, entonces para analizar los errores completamente necesitaremos las cotas para las cantidades 
  $$\frac{|\hat{(\hat x \diamondsuit \hat y)}-(x\diamondsuit y) |}{|(x\diamondsuit y)|}$$  

  Estas cotas varian de acuerdo  a la naturaleza de la operación
